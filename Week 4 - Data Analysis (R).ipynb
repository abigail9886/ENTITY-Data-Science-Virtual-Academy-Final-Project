# Run a Multiple Linear Regression on the Continuous Variables

```r
# Remove `instant`, `season`, `yr`, `mnth`, `hr`, `holiday`, `weekday`, `workingday`, `weathersit`, `casual`, `registered` so we are left with only the continuous predictor variables with `cnt` as DV
bikes1 <- subset(bikes, select = -c(instant, season, yr, mnth, hr, holiday, weekday, workingday, weathersit, casual, registered))
head(bikes1, 10)
```

```r
bikes2 <- bikes1

# Coerce `dteday` to date
bikes2$dteday <- as.Date(bikes2$dteday)

# Coerce `
head(bikes2, 10)
```

## Backward Regression

```r
# Run a linear regression model with all possible variables included
FitAll = lm(cnt ~ ., data = bikes2)
summary(FitAll)
```

```r
# Run a backwards regression
step(FitAll, direction = 'backward')
```

```r
fitsome <- lm(cnt ~ dteday + temp + atemp + hum + windspeed, data = bikes2)
summary(fitsome)
```

## Forward Regression

```r
fitstart <- lm(cnt ~ 1, data = bikes2)
summary(fitstart)
```

```r
step(fitstart, direction = 'forward', scope = (formula(FitAll)))
```

Both forward and backward regression have shown that all continuous variables in the dataset are good predictors of the `cnt` dependent variable.

# Run a MANOVA on the Categorical Variables with all 3 possible DVs

```r
# Make a new copy of the original dataset
bikes3 <- bikes
head(bikes3, 10)
```

```r
# Subset to only the categorical IVs and the 3 possible DVs
bikes4 <- subset(bikes3, select = -c(instant, dteday, temp, atemp, hum, windspeed))
head(bikes4, 10)
```

```r
# Coerce the categorical variables to factors
bikes4$season <- as.factor(bikes4$season)
bikes4$yr <- as.factor(bikes4$yr)
bikes4$mnth <- as.factor(bikes4$mnth)
bikes4$hr <- as.factor(bikes4$hr)
bikes4$holiday <- as.factor(bikes4$holiday)
bikes4$weekday <- as.factor(bikes4$weekday)
bikes4$workingday <- as.factor(bikes4$workingday)
bikes4$weathersit <- as.factor(bikes4$weathersit)

bikes4
```

## Test for Assumptions for MANOVA

### Sample Size
"There must be more cases than dependent variables in every cell. In addition, there must be at least 20 cases per independent variable."

Given that there are 17,379 observations, we know that we far exceed the 20 cases per independent variable. With 8 IVs, we would need a minimum of 160 cases, so we far exceed that minimum.

```r
table(bikes4$season)
table(bikes4$yr)
table(bikes4$mnth)
table(bikes4$hr)
table(bikes4$holiday)
table(bikes4$weekday)
table(bikes4$workingday)
table(bikes4$weathersit)
```

Nearly every cell of the table counts has more than the minimum of 3 required for the 3 DVs we will be testing. The only exception is the 3 (equal, rather than greater) in the `4` of `weathersit`. We will continue with our analysis anyway.

### Multivariate Normality
We begin by subsetting to only our DVs.

```r
bikes4dv <- subset(bikes4, select = c(casual, registered, cnt))
bikes4dv
```

```r
# Limit number of rows because the Wilks-Shapiro test cannot take too much data
bikes4dv1 <- bikes4dv[1:5000,]
bikes4dv1
```

```r
# Run Wilks-Shapiro test for Multivariate Normality
mshapiro.test(t(bikes4dv1))
```

We do NOT meet the assumption of multivariate normality that is required for MANOVA. However, we were never taught what to actually do in that case, only to continue on for the sake of learning. So, that is what we will do.

### Homogeneity of Variance

We have previously discovered that our DVs are not normally distributed. Therefore, we will use the Fligner-Killeen test for homogeniety of variance, as it is known to be the most forgiving option when dealing with non-normal data.

Although we have 8 possible categorical IVs, I have chosen only 3 to check below as if 3 are not homogenous in variance, then the full 8 certainly will not be either.

```r
fligner.test(casual ~ interaction(season, weekday, holiday), data = bikes4)
```

Unfortunately, the `casual` DV does not meet the assumption of homogeneity of variance. Therefore, we expect that `registered` and `cnt` will not meet it either, as we already know that the 3 DVs are closely related. However, we can check them quickly as well.

```r
fligner.test(registered ~ interaction(season, weekday, holiday), data = bikes4)
fligner.test(cnt ~ interaction(season, weekday, holiday), data = bikes4)
```

Indeed, we see that these data do NOT meet the assumption of homogeniety of variance in order to proceed with the MANOVA. However, for exploration purposes, we will continue with the knowledge that our final analysis may not be completely reliable.

### Absence of Multicollinearity
We run a correlation on all combinations of the DVs.

```r
cor.test(bikes4dv$casual, bikes4dv$registered, method = 'pearson', use = 'complete.obs')
cor.test(bikes4dv$casual, bikes4dv$cnt, method = 'pearson', use = 'complete.obs')
cor.test(bikes4dv$registered, bikes4dv$cnt, method = 'pearson', use = 'complete.obs')
```

We can see that only `casual` and `registered` meet the assumption of the absence of multicollinearity. Indeed, `casual` is very close to the 0.7 boundary of multicollinearity with `cnt`, while `registered` is extremely closely correlated with `cnt`. Because of these relationships, we will use only the `casual` and `registered` variables to run the MANOVA. (In other words, `registered` would account for nearly all of the variance in `cnt` as well, so there is no need to include both variables.)

## Run the MANOVA

```r
ind_vars <- cbind(bikes4$season, bikes4$yr, bikes4$mnth, bikes4$hr, bikes4$holiday, bikes4$workingday, bikes4$weekday, bikes4$weathersit)

MANOVA <- manova(cbind(casual, registered) ~ ind_vars, data = bikes4)
summary(MANOVA)
```

```r
summary.aov(MANOVA, test = 'wilks')
```

We see from the post-hocs that the independent variables are significant predictors for both `casual` and `registered` users.
